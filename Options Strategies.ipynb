## Options Strategies

### 1) Protective Put

```python
def plot_strategies(K):
    St = np.linspace(0.5 * K, 1.5 * K, 100)
    
    protective_put_payoff = np.maximum(K, St)
    protective_put_pl = np.maximum(protective_put_payoff - 115, 0)
    
    fig, axes = plt.subplots(2, 2, figsize=(10, 6))  # Create a 2x2 grid of subplots
    
    # Protective Put Payoff
    axes[0, 0].plot(St, protective_put_payoff, label='Protective Put', color='darkturquoise')
    axes[0, 0].legend()
    
    # Protective Put P/L
    axes[0, 0].plot(St, protective_put_pl, label='Protective Put P/L', color='salmon')
    axes[0, 0].set_title('Protective Put P/L')
    axes[0, 0].legend()
    axes[0, 0].set_title('Protective Put Payoff & P/L')
    axes[0, 0].grid()

# Parameters
K = 100  # Strike price
plot_strategies(K)
```
